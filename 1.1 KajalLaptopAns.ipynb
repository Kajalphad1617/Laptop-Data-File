```python
import pandas as pd
df=pd.read_csv("laptop_price.csv")
df
```

```python
#Q1.Count of each laptop according their brand in given dataset.
brand_counts = df['Company'].value_counts()
print(brand_counts)
```

```python
#Q2.Bar graph which represent Operating sys. 
import matplotlib.pyplot as plt
osc = df['OpSys'].value_counts()
osc.plot(kind='bar', color='skyblue')
plt.title('Count of Laptops by Operating System')
plt.xlabel('Operating System')
plt.ylabel('Count')
plt.grid(axis='y', linestyle='--')
plt.show()
```
